# Visualization with Matplotlib

Loading the basic packages required for this exercise:

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Data Visualization (Linechart plot) in Matplotlib

```python
x = np.linspace(0,10,30)
```

```python
y = np.sin(x)
```

### Plotting with plt.plot()

```python
plt.plot(x,y)
```

### Adding Additional Elements to the Plot

Adding the x-axis label to the plot:

```python
plt.plot(x,y)
plt.xlabel("Time")
```

Adding the y-axis label to the plot:

```python
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Some function of time")
```

Adding a title to the plot:

```python
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("My cool chart")
```

**CONCLUSION:** The .pyplot function works very much like the basic plot packages in R - in that in functions in an additive manner, e.g.: 
- the first line specifies the objects to be plotted,
- the second line **adds** an x-axis label,
- the third line **adds** a y-axis label, and
- the fourth line **adds** a title to the plot

**NOTE:** Given the many other possible variations to this function, try doing an exploration of this function on your own, e.g.: changing line colors, adding grids, adding text to the lines, changing line types, etc

Scatterplot **only(?)** works when the dataframe is converted into a matrix - in order to work with the .scatter function in this exercise, we need to convert the loaded dataset (dataframe) into a numpy matrix

```python
table_matrix = table.as_matrix()
```

**Pay attention to the syntax to select specific values from a matrix**

```python
x = table_matrix[:,0]
y = table_matrix[:,1]
plt.scatter(x,y)
```

Just like the .plot function in the previous section, the scatterplot also works in an additive fashion. Such that addition to the plot can be made in the following manner:

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1
```

The above code creates a line to be **added into the plot** - in this case, we just happens to know what the line is

```python
plt.scatter(x,y)
plt.plot(x_line, y_line) #Adding a line to the plot
plt.xlabel("Time") #Adding the label "Time" to the plot
plt.ylabel("Some function of time") #Adding the label "Some function of time" to the plot
plt.title("My cool chart") #Adding the title "My cool chart" to the plot
```

## Visualizing Histograms

**NOTE:** A histogram is essentially just a plot that shows the frequency of observations for every bucket of values

```python
plt.hist(x)
```

Creating a random value of 10000 observations:

```python
R = np.random.random(10000)
plt.hist(R)
```

By the default, the .hist function divides values into 10 brackets. The **bins** argument allows us to manually define the number of brackets the histogram should divide the value into.

```python
plt.hist(R, bins = 20)
```

Imagine the case where we're doing a linear regression: we have the actual y values, the predicted y, and the residuals (difference) between the two values:

```python
y_actual = 2*x - 1
residuals = y - y_actual
```

Plotting the residual values in a histogram, gives us:

```python
plt.hist(residuals)
```

The form above more or less forms a bell-shaped curve, but given that we only have 100 observations for 'x', the bell-shape may not be readily apparent

## Plotting Images

When calling an image to a plot, consider the following features that is pertinent to how images are being stored in a computer:
- an image is essentially **a matrix of numbers**
- A(i,j) represents the pixel intensity at coordinate (i,j)
- JPG or PNG are not matrices, since they are compressed
- to get the matrix of values for a JPG or PNG images, we'll need to decompress these images

For this next example, we'll be using the MNIST dataset (handwritten digits) from [Kaggle](https://kaggle.com/c/digit-recognizer) - or https://www.kaggle.com/c/digit-recognizer/data

```python
import pandas as pd
data = pd.read_csv("./machine_learning_examples-master/large_data/train.csv")
```

```python
data.info()
```

As before, matplotlib **only(?)** works with numpy arrays and matrices - so in order to get this to work, we'll need to convert the loaded dataframe into a matrix

```python
data_matrix = data.as_matrix()
```

Subsetting the first row of the dataset:

```python
im = data_matrix[0,1:]
im.shape
```

```python
im
```

The above script somewhat represents an image (0 for no image, value 1 - 255 represents differnt intensity of colors). However, to actually plot the image, we'll need to reshape this row of data into an X by X grid format, following an image's storage format (essentially a matrix, with each of its values representing the 'intensity' of its associated cells)

To **convert the row into a matrix**:

```python
im = im.reshape(28,28)
im.shape
```

### Plotting an Image

**Basic Image Plotting**

```python
plt.imshow(im)
```

**Plotting an Image in Black and White**

```python
plt.imshow(im, cmap = 'gray')
```

**Inverting the black and white of the Image**

```python
plt.imshow(255 - im, cmap = 'gray')
```
